```python
#creat a new dataframe containing the needed columns only
df1 = df[['SK_ID_CURR',
          'AMT_ANNUITY',
        'AMT_APPLICATION' ,
           'AMT_CREDIT',
        'AMT_DOWN_PAYMENT',
         'AMT_GOODS_PRICE',
          'DAYS_DECISION',
       'HOUR_APPR_PROCESS_START',
      'RATE_DOWN_PAYMENT',
        'SK_ID_PREV']]
print(df1.shape)
```

```python
# Aggregation for numeric features
num_aggregations = {
        'SK_ID_PREV' : 'count',
        'AMT_ANNUITY': ['min', 'max', 'median', 'mean'], 
        'AMT_APPLICATION': ['min', 'max', 'mean', 'median'],
        'AMT_CREDIT': ['min', 'max', 'mean', 'median'],
        'AMT_DOWN_PAYMENT': ['min', 'max', 'mean', 'median'],
        'AMT_GOODS_PRICE': ['min', 'max', 'mean', 'median'],
        'DAYS_DECISION': ['min', 'max', 'mean','median'],
        'HOUR_APPR_PROCESS_START': ['min', 'max', 'mean'],
        'RATE_DOWN_PAYMENT': ['min', 'max', 'mean','median'],
               
    }
```

```python
#creat a new dataframe containing the needed aggregation 
Final_results_python = df1.groupby(by='SK_ID_CURR').agg({**num_aggregations})
```

# SQL steps to produce the requested

```python
#imported the previous applications file into an SQLite database to be used in this part

#imports
import sqlite3
import pandas as pd


#initiate connection to database
connection = sqlite3.connect('/home/khaldoun/Desktop/MBDS/Data-driven Business/3rd Assigment /p_db2.db')
cursor = connection.cursor()


#test data retrival  to make sure connection is working properly 
cursor.execute("SELECT * from 'previous_application'LIMIT 4 ")
connection.commit()
rows = cursor.fetchall()
pd.DataFrame(rows).head()
```

```python
# prepare query to retrive needed data and do aggregation 


Query_part1 = "select SK_ID_CURR,count(sk_id_prev) AS CntID_Prev,"

Query_part2 = "min(AMT_ANNUITY) ,max(AMT_ANNUITY) ,AVG(AMT_ANNUITY) ,"

Query_part3 = "min(AMT_APPLICATION) ,max(AMT_APPLICATION),AVG(AMT_APPLICATION),"

Query_part4 = "min(AMT_CREDIT),max(AMT_CREDIT),AVG(AMT_CREDIT),"

Query_part5 = "min(AMT_DOWN_PAYMENT) ,max(AMT_DOWN_PAYMENT),AVG(AMT_DOWN_PAYMENT),"

Query_part6 = "min(AMT_GOODS_PRICE) ,max(AMT_GOODS_PRICE),AVG(AMT_GOODS_PRICE),"

Query_part7 = "min(DAYS_DECISION) ,max(DAYS_DECISION),AVG(DAYS_DECISION),"

Query_part8 = "min(HOUR_APPR_PROCESS_START) ,max(HOUR_APPR_PROCESS_START),AVG(HOUR_APPR_PROCESS_START),"

Query_part9 = "min(RATE_DOWN_PAYMENT) ,max(RATE_DOWN_PAYMENT),AVG(RATE_DOWN_PAYMENT)"

Query_part10 = " from previous_application group by SK_ID_CURR"

Main_query =  str(Query_part1 + Query_part2 + Query_part3 + Query_part4 + Query_part5 + Query_part6 + Query_part7 + Query_part8 +Query_part9 + Query_part10)

print(Main_query)
```

```python
# Excute query 

cursor.execute(Main_query)
connection.commit()
query_results = cursor.fetchall()


#load query results into pandas dataframe
Final_results_SQL = pd.DataFrame(query_results  , columns = ['SK_ID_CURR' , 'count(sk_id_prev)' , 
                                         'min(AMT_ANNUITY) ' ,'max(AMT_ANNUITY)' , 'AVG(AMT_ANNUITY)',
                                        'min(AMT_APPLICATION)','max(AMT_APPLICATION)' ,'AVG(AMT_APPLICATION)',
                                        'min(AMT_CREDIT)', 'max(AMT_CREDIT)','AVG(AMT_CREDIT)',
                                        'min(AMT_DOWN_PAYMENT)','max(AMT_DOWN_PAYMENT)','AVG(AMT_DOWN_PAYMENT)',
                                        'min(AMT_GOODS_PRICE)', 'max(AMT_GOODS_PRICE)','AVG(AMT_GOODS_PRICE)',
                                        'min(DAYS_DECISION)', 'max(DAYS_DECISION)','AVG(AMT_CREDIT)',
                                        'min(HOUR_APPR_PROCESS_START)', 'max(HOUR_APPR_PROCESS_START)','AVG(HOUR_APPR_PROCESS_START)',
                                        'min(RATE_DOWN_PAYMENT)', 'max(RATE_DOWN_PAYMENT)','AVG(RATE_DOWN_PAYMENT)'])

# close connection
connection.close()
```
